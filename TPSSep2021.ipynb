```python
#imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
train = pd.read_csv('/content/gdrive/MyDrive/achyut.kr.de/Coder/Kaggle/02 TPSSep 2021/data/train.csv')
test = pd.read_csv('/content/gdrive/MyDrive/achyut.kr.de/Coder/Kaggle/02 TPSSep 2021/data/test.csv')

train = train.fillna(method='ffill')
train = train.dropna()

test = test.fillna(method='ffill')
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(train.iloc[:,1:-1])
X_test = sc.transform(test.iloc[:,1:])
y_train = train.iloc[:,-1:]
```

```python
from xgboost import XGBRegressor
regressor = XGBRegressor(criterion = 'entropy', verbose = 100)
regressor.fit(X_train,y_train)

y_pred = regressor.predict(X_test)
```

```python
ans = test.iloc[:,:1].values.tolist()
answer = []
for i in range(len(test)):
  answer.append(ans[i][0])
```

```python
output = pd.DataFrame({'id': answer, 'claim': y_pred})
output.to_csv('/content/gdrive/MyDrive/achyut.kr.de/Coder/Kaggle/02 TPSSep 2021/data/answer.csv', index=False)
print("Your submission was successfully saved!")
```
